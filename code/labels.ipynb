# LABELS IN MATPLOTLIB

Labels describe what your data represents, so graphs are readable and meaningful.

- They answer:
    1. X-axis → What is this?
    2. Y-axis → What is this?
    3. Line/plot → What does this line represent?

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Most Important Basic Functions

1. `plt.title`:
Gives context to your graph.

Parameters:
- `fontsize=''`
-  font family(`family`)
- `fontweight=` 

```python
x = np.array([2020,2021,2022,2023,2024,2025])  #YEARS
y = np.array([30, 33, 36, 42, 45, 50])  #students in class
plt.plot(x, y)  

plt.title('Student Per Year', fontsize='20', family='sans-serif', fontweight='bold', color='red')
plt.show()
```

2. **`plt.xlabel()` & `plt.ylabel()`**: Explain axes clearly (this is professional practice).

```python
x = np.array([2020,2021,2022,2023,2024,2025])  #YEARS
y = np.array([30, 33, 36, 42, 45, 50])  #students in class
plt.plot(x, y)  
plt.title('Student Per Year', fontsize='20', family='sans-serif', fontweight='bold', color='#00de21')


plt.xlabel('Years', fontsize='10', family='sans-serif', fontweight='bold', color='#1261e0')
plt.ylabel('Students', fontsize='10', family='sans-serif', fontweight='bold', color="#FF0000")
plt.show()
```

**`plt.xticks()` & `plt.yticks()`**: Ticks are the numbers/labels shown on the axes that help you read values.

- X-ticks → values on X-axis
- Y-ticks → values on Y-axis

**`plt.tick_params()`**: Controls the appearance of ticks.


```python
x = np.array([2020,2021,2022,2023,2024,2025])  #YEARS
y = np.array([30, 33, 36, 42, 45, 50])  #students in class
plt.plot(x, y)  
plt.title('Student Per Year', fontsize='20', family='sans-serif', fontweight='bold', color='#00de21')


plt.xlabel('Years', fontsize='10', family='sans-serif', fontweight='bold', color='#1261e0')
plt.ylabel('Students', fontsize='10', family='sans-serif', fontweight='bold', color="#FF0000")

plt.tick_params(axis='x', colors='#96c40a')
plt.xticks(x)
plt.show()
```

3. **`plt.figure(figsize=('width'inches,'height'inches))`**: Controls size of the plot
- Real-world dashboards always control figure size.
- Common Parameters: figsize, dpi, edgecolor, num.
- Each `plt.figure()` creates a new figure.

Key Rules to Remember!

- ⚠️ Call plt.figure() only once per plot
- All styling must be set on the same figure
- Multiple plt.figure() = multiple separate canvases

simply plt.figure should be called once per cell. and should be before plt.plot(x,y) so that it can be applied.

```python
x = np.array([2020,2021,2022,2023,2024,2025])  #YEARS
y = np.array([30, 33, 36, 42, 45, 50])  #students in class

plt.figure(figsize=(7,5), linewidth=5, edgecolor='white', facecolor='red')

plt.plot(x, y)  
plt.title('Student Per Year')
plt.xlabel('Years')
plt.ylabel('Students')
plt.grid(True)
plt.show()
```
