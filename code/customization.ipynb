# CUSTOMIZATION IN MATPLOTLIB

Customizing plots, lines and markers.

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### **`marker=''`** arguement in plot function.
- marker controls the shape of points on a line or scatter plot.
- marker='given character'. For e.g: marker='o'.
This draws circles at each data point.

#### Common marker types
- `o`   ==	Circle
- `.`	==    Point
- `s`   ==	Square
- `^`   ==	Triangle up
- `v`	==    Triangle down
- `D`	==    Diamond
- `x`	==    X mark
- `+`	==    Plus
- `*`	==    Star
- `None` or `''`	  ==  No marker

#### Customization Options
You can further customize markers using additional parameters in your plotting functions:
- `markersize=''` (or `ms`): Adjusts the size of the marker.
- `markerfacecolor=''` (or `mfc`): Sets the fill color of the marker.
- `markeredgecolor=''` (or `mec`): Sets the color of the marker's edge.
- `markeredgewidth=''` (or `mew`): Controls the thickness of the marker's edge.
- `fillstyle`: Allows for options like 'full', 'left', 'right', 'bottom', 'top', or 'none' for partial or no filling of the marker. 

```python
x = np.array([2020,2021,2022,2023,2024,2025])  #YEARS
y = np.array([30, 33, 36, 42, 45, 50])  #students in class

plt.plot(x, y, 
         marker='v',
         markersize=10,
         markerfacecolor='#ab1b1beb',#HSL VALUE IS USED  
         markeredgecolor='black')   

plt.show()
```

#### **`linestyle=''`** arguement in plot function.
- You can select line style, the default value is *solid*.
- linestyle controls how the line looks (solid, dashed, dotted, etc.).
- `linestyle=''`/ `ls`.

#### Common linestyles

- Solid ==	`-`
- Dashed    ==	`--`
- Dotted    ==	`:`
- Dash-dot	==    `-.`
- No line  ==	`None`

#### Real-world usage
- Solid (-) → main trend
- Dashed (--) → comparison / forecast
- Dotted (:) → reference / baseline

```python
x = np.array([2020,2021,2022,2023,2024,2025])  #YEARS
y = np.array([30, 33, 36, 42, 45, 50])  #students in class

plt.plot(x, y, 
         marker='o',
         markersize=10,
         markerfacecolor='#ab1b1beb',#HSL VALUE IS USED  
         markeredgecolor='black',
         
         linestyle='--')   

plt.show()
```

#### **`linewidth=''`** arguement in plot function
- Is a property of a line plot that controls the thickness of the drawn line. 
- The default value for linewidth is typically 1.5 points
- `linewidth=''`/`lw`

```python
x = np.array([2020,2021,2022,2023,2024,2025])  #YEARS
y = np.array([30, 33, 36, 42, 45, 50])  #students in class

plt.plot(x, y, 
         marker='o',
         markersize=10,
         markerfacecolor='#ab1b1beb',#HSL VALUE IS USED  
         markeredgecolor='black',
         linestyle='-',
         
         linewidth=3)   

plt.show()
```

#### **`color=''`** arguement in plot function.
- Controls line color
- Takes Common values: 'r', 'g', 'b' | 'black', 'green' | Hex: '#FF5733'

```python
x = np.array([2020,2021,2022,2023,2024,2025])  #YEARS
y = np.array([30, 33, 36, 42, 45, 50])  #students in class

plt.plot(x, y, 
         marker='o',
         markersize=10,
         markerfacecolor='#ab1b1beb',#HSL VALUE IS USED  
         markeredgecolor='black',
         linestyle='-',
         linewidth=3,
         
         color= '#4169E1')  

plt.show()
```

#### MULTIPLE LINE CUSTOMIZATION WITH HELP OF DICTIONARY()
- Instead of passing individually custom arguments(line width, color etc), we can call a dictionary
- To call a dictionary in matplotlib, use ` ** `double star to call it, otherwise it will through error.

```python
x = np.array([2019, 2020, 2021, 2022, 2023, 2024])  #years
y = np.array([28, 30, 34, 38, 42, 47])  #students
y1 = np.array([25, 28, 32, 35, 40, 45])
y2 = np.array([34,32,41,23,46,8])



line_style_dictionary= dict(marker='o',
                       markersize=10,
                       markerfacecolor='#ab1b1beb',#HSL VALUE IS USED  
                       markeredgecolor='black',
                       linestyle='-',
                       linewidth=3,
                      )

plt.plot(x, y,  color= '#4169E1',   **line_style_dictionary)
plt.plot(x, y1, color='red',    **line_style_dictionary)
plt.plot(x, y2, color='green',  **line_style_dictionary)

plt.show()
```
