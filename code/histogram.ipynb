# Histogram in Matplotlib
- A **`Histogram`** is a Visual representation of the distribution of quantitative data.
- It groups values into bins (intervals) and counts how many fall in each range.
- Great for Statistics and Probability.
- In many cases, Bell Curve distribution is seen(like this == `_/\_`).
- Syntax: **`plt.hist(data)`**



```python
import numpy as np
import matplotlib.pyplot as plt
```

- Here we used `np.random.normal(loc=, scale=, size=)` function...
- The np.random.normal() function in NumPy draws random samples from a normal (Gaussian) distribution, often referred to as the "bell curve". This distribution is widely used in statistics and data analysis because it models many natural phenomena. 
- Parameters:
    1. loc: (Mean) The central peak of the distribution. The default value is 0.0.
    2. scale: (Standard Deviation) Controls the spread or "width" of the distribution. Must be non-negative. The default value is 1.0.
    3. size: (Output Shape) An integer or tuple of integers specifying the shape of the output array. If None (default), a single scalar value is returned. 
- We used np.clip just to set boundries so that score wont go below 0 and not more than 100.

```python
student_scores_array = np.random.normal(loc=80, scale=10, size=100) 
student_scores_array = np.clip(student_scores_array, 0, 100) 

plt.hist(student_scores_array, 
         bins=20,
         color='#0e0087',
         edgecolor='red')

plt.title("Exam Scores")
plt.xlabel("Score")
plt.ylabel("Number of Students")


plt.show()
```
