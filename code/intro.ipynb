```python
x_coordinates = [2020,2021,2022,2023,2024,2025]  #YEARS
y_coordinates = [30, 33, 36, 42, 45, 50]  #students in class

plt.plot(x_coordinates, y_coordinates)  

plt.show()
```

IMPORTING NUMPY AS ARRAY INSTEAD OF PYTHON LIST BECAUSE OF SPEED

```python
import numpy as np

x = np.array([2020,2021,2022,2023,2024,2025])  #YEARS
y = np.array([30, 33, 36, 42, 45, 50])  #students in class

plt.plot(x, y)  

plt.show()
```
