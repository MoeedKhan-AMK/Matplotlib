# Creating a graph using Pandas and Matplotlib
- In this Program, We are integrating Pandas and Matplotlib. This will be beneficial for as a beginner in matplotlib.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('players_18.csv')
df
```

- Creating Bar chart to show club name

```python
club_position_count = df['club_position'].value_counts().head(10)
club_position_count
```

```python
club_position_count = df['club_position'].value_counts(ascending=True).head(10) #ascending=True means min to max.

plt.barh(club_position_count.index, club_position_count.values)
plt.title('Number of Club Positions')
plt.xlabel('Count')
plt.ylabel('Positions')

plt.show()
```
