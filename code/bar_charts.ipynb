# BAR CHARTS IN MATPLOTLIB

- **Bar Charts:** A visual tool that uses rectangular bars to compare different categories of data, where the length or height of each bar is proportional to the value it represents, making it easy to see comparisons and trends at a glance.
![bar chart example](barchart.png)

We are creating bar chart to represent daily consumption of categories of foods using food pyramid and bar chart.
![alt text](food-pyramid.png)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
food_categories = np.array(['Whole Grains',' Legumes', 'Fruits','Protien' ,'Vegetables', 'Sweets'])
serving_value = np.array([5,2,3,7,4,1])
```

### `plt.bar()`: create vertical bar charts

```python
plt.bar(food_categories, serving_value, color='Green')

plt.title('Daily Consumption')
plt.xlabel('Food')
plt.ylabel('Serving Quantity')

plt.show()
```

### `plt.barh()`:  create horizontal bar charts.


```python
plt.barh(food_categories, serving_value, color='skyblue')

plt.title('Daily Consumption')
plt.xlabel('Food')
plt.ylabel('Serving Quantity')

plt.show()
```
