# GRID() IN MATPLOTLIB

**`plt.grid()`**: Used to toggle the visibility of grid lines on a plot, which enhances readability(especially for clients). 
- It adds reference lines which makes plots to easier to read. 
![grid()](plot_graph.png)
- we can set axis of grid lines: `plt.grid(axis='x or y')
- You can change style, width color etc for grid lines.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([2020,2021,2022,2023,2024,2025])  #YEARS
y = np.array([30, 33, 36, 42, 45, 50])  #students in class

plt.figure(figsize=(7,5), linewidth=5, edgecolor='white', facecolor='#1261e0')
plt.plot(x, y)  
plt.title('Student Per Year')
plt.xlabel('Years')
plt.ylabel('Students')

plt.grid(axis='both',
         color='red')

plt.show()
```
