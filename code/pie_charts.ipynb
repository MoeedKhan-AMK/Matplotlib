# PIE CHARTS IN MATPLOTLIB

- **Pie Chart:** a type of graph representing data in a circular form, with each slice of the circle representing a fraction or proportionate part of the whole.
- Good for visualizing distribution among categories.
![pie chart](pie_chart.png)

### `plt.pie(value, label='labels')`
**Important PARAMETERS**
- `labels`: A list of strings providing labels for each slice.
- `autopct`: auto percentage. set `autpct='%1.1f%%'` to get percentage of slice.
-  You can set slice wise color: `colors= list of color`
- *plt.axis('equal') is recommended to prevent the chart from appearing as an oval.* 


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
student_categories_labels = np.array(["Freshmans", "Sophomores", "Juniors", "Seniors"])
student_values = np.array([433,300,320,156])
slice_colors = ["#3198A7", "#15616D", "#FF7D00", "#78290F"]

plt.pie(student_values, labels=student_categories_labels, autopct='%1.1f%%', colors=slice_colors, startangle=90)
plt.axis('equal')

plt.title("Students Percentage Per Category")
plt.show()
```

#### `Explode` Parameter:
An array of values that specify the fraction of the radius to offset a slice, used to highlight a specific portion of the data.
Simply It will cutt off a slice as shown in figure: 



- It takes List as arguement.
- `plt.pie(explode=[values per labels])`

```python
student_categories_labels = np.array(["Freshmans", "Sophomores", "Juniors", "Seniors"])
student_values = np.array([433,300,320,156])
slice_colors = ["#3198A7", "#15616D", "#FF7D00", "#78290F"]

plt.pie(student_values, labels=student_categories_labels, autopct='%1.1f%%', colors=slice_colors, explode=[0,0,0,0.1])
plt.axis('equal')

plt.title("Students Percentage Per Category")
plt.show()
```

#### `startangle=` Parameter:
The angle at which the first slice is drawn, rotated counterclockwise from the positive x-axis (default is 0 degrees, 90 degrees starts at the top).

```python
student_categories_labels = np.array(["Freshmans", "Sophomores", "Juniors", "Seniors"])
student_values = np.array([433,300,320,156])
slice_colors = ["#3198A7", "#15616D", "#FF7D00", "#78290F"]

plt.pie(student_values, labels=student_categories_labels, autopct='%1.1f%%', colors=slice_colors, startangle=90)
plt.axis('equal')

plt.title("Students Percentage Per Category")
plt.show()
```

#### `shadow=true` PARAMETER: 
Drop Shadow Will be added.


```python
student_categories_labels = np.array(["Freshmans", "Sophomores", "Juniors", "Seniors"])
student_values = np.array([433,300,320,156])
slice_colors = ["#3198A7", "#15616D", "#FF7D00", "#78290F"]

plt.pie(student_values, labels=student_categories_labels, autopct='%1.1f%%', colors=slice_colors, shadow=True)
plt.axis('equal')

plt.title("Students Percentage Per Category")
plt.show()
```
