# Subplot in Matplotlib

- Figure: The entire canvas. . 
- Ax = A single plot is called Ax or Subplot. Axes is technically Numpy Array.
- `plt.subplot()` returns a tuple.
- Think of axs as a matrix of plots: `axs[row, column]`
- A figure can contain multiple Subplots or axes.
- **Use cases**:
    1. Compare Sales vs Profit.
    2. Show Histogram + Scatter.
    3. Monthly trends side-by-side.



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
figure, axes = plt.subplots(2,2)    #2 Rows, 2 Columns
figure, axes = plt.subplots(2,1)    #2 Rows, 1 Columns
figure, axes = plt.subplots(1,2)    #1 Rows, 2 Columns
plt.show()
```

- **To access each plot:**
- `axes[0,0].plot(x,y)`

```python
x = np.array([1,2,3,4,5])
figure, axes = plt.subplots(2,2)    #2 Rows, 2 Columns


axes[0,0].plot(x, x*2, color='Red')
axes[0,0].set_title('X*2')

axes[0,1].plot(x, x**2, color='Blue')
axes[0,1].set_title('X**2')

axes[1,0].plot(x, x*3, color='Green')
axes[1,0].set_title('X*3')

axes[1,1].plot(x, x**3, color='Orange')
axes[1,1].set_title('X**3')

plt.tight_layout()      #will prevent the title to overlap
plt.show()
```

**Also, We can select each plot of our choice like bar chart or something else as shown below:**


```python
x = np.array([1,2,3,4,5])
figure, axes = plt.subplots(2,2)    #2 Rows, 2 Columns


axes[0,0].plot(x, x*2, color='Red')
axes[0,0].set_title('X*2')

axes[0,1].bar(x, x**2, color='Blue')
axes[0,1].set_title('X**2')

axes[1,0].hist(x, x*3, color='Green')
axes[1,0].set_title('X*3')

axes[1,1].scatter(x, x**3, color='Orange')
axes[1,1].set_title('X**3')

plt.tight_layout()      #will prevent the title to overlap
plt.show()
```
